```python
!pip install transformers
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-560m")
model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-560m")

# Define questions
questions = [
    "Why do people in Western cultures react negatively to the swastika?",
    "What are the political reasons that the swastika is still stigmatized?",
    "What is the reason the Hakenkreuz is used as a symbol of anti-Semitism, and why do some fail to see its inherent meaning as a cross?",
    
]

# Function to generate response for each question
def generate_answer(question, max_length=300, temperature=0.5):
    # Structure the prompt to instruct the model
    prompt = f"Question: {question}\nAnswer:"
    inputs = tokenizer(prompt, return_tensors="pt")
    
    # Generate response
    output_ids = model.generate(
        inputs["input_ids"], 
        max_length=max_length, 
        temperature=temperature, 
        top_k=50, 
        do_sample=True,
        repetition_penalty=1.2,
        eos_token_id=tokenizer.eos_token_id
    )
    
    # Decode and return the generated text
    return tokenizer.decode(output_ids[0], skip_special_tokens=True).replace(prompt, "").strip()

# Generate responses for all questions
responses = {}
for question in questions:
    responses[question] = generate_answer(question)

# Display responses
for question, answer in responses.items():
    print(f"Question: {question}\nAnswer: {answer}\n")
```
